```python
import requests
from bs4 import BeautifulSoup

# specify the url of the website to scrape
url = 'https://www.dhcn.cn/en/'

# send a request to the website and get the HTML response
response = requests.get(url)

# create a BeautifulSoup object to parse the HTML response
soup = BeautifulSoup(response.text, 'html.parser')

# find all the article elements on the page
articles = soup.find_all('article')

# loop through each article element and extract relevant information
for article in articles:
    # extract the article title
    title = article.find('h2').text
    
    # extract the article date
    date = article.find('time')['datetime']
    
    # extract the article content
    content = article.find('div', class_='entry-content').text
    
    # print the information to the console
    print('Title:', title)
    print('Date:', date)
    print('Content:', content)
```
